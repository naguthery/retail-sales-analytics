#### Import Libraries

```python
import pandas as pd
import numpy as np

# Display all columns
pd.set_option('display.max_columns', None)

# Display all rows
pd.set_option('display.max_rows', None)
```

#### Load Dataset

```python
df = pd.read_csv('synthetic_online_retail_data.csv')

# Return first 5 rows to make sure the dataset has been loaded correctly.
df.head()
```

#### Dataset Inpection
Examine the stuructre and characterisitcs of the dataset.

```python
# Dataset Dimensions
print("Dataset Dimensions")
print("Rows and Columns: ", df.shape)

print("")

# Column names
print("Column Names")
print(df.columns)

print("")

# Data types
print("Data Types")
print(df.info())

print("")

# Summary statistics
print("Summary Statistics")
print(df.describe())
```

#### Identify Issues in the Datset

```python
# Check for missing values
print("Missing Values")
print(df.isnull().sum())

print("")

# Check for duplicate values
print("Duplicate Values")
print(df.duplicated().sum())

print("")

# Check data types
print("Date Types")
print(df.dtypes)

# Convert order_date column
df['order_date'] = pd.to_datetime(df['order_date'])
```

#### Data Cleaning
Fill in the missing values

```python
# Fill in mising values for review_score
df['review_score'] = df['review_score'].fillna(0)

# Fill in missing values for gender
df['gender'] = df['gender'].fillna('Unknown')
```

#### Verify Changes Have Been Made

```python
# Check for missing values
print("Missing Values")
print(df.isnull().sum())

print("")

# Check for duplicate values
print("Duplicate Values")
print(df.duplicated().sum())

print("")

# Check data types
print("Data Types")
print(df.dtypes)
```

#### Final Dataset Inspection

```python
# Dataset Dimensions
print("Dataset Dimensions")
print("Rows and Columns: ", df.shape)

print("")

# Column names
print("Column Names")
print(df.columns)

print("")

# Data types
print("Data Types")
print(df.info())

print("")

# Summary statistics
print("Summary Statistics")
print(df.describe())
```
